```python
# ‚úÖ –Ø–ß–ï–ô–ö–ê 2: –°–æ–∑–¥–∞–Ω–∏–µ –∞–≥–µ–Ω—Ç–æ–≤ (–ò–°–ü–†–ê–í–õ–ï–ù–ù–ê–Ø –í–ï–†–°–ò–Ø)
import asyncio
import nest_asyncio
import warnings
import logging

# –ü–æ–¥–∞–≤–ª—è–µ–º warning'–∏ –¥–ª—è —á–∏—Å—Ç–æ–≥–æ –≤—ã–≤–æ–¥–∞
warnings.filterwarnings('ignore')
logging.getLogger('agents.operational.lead_qualification').setLevel(logging.ERROR)
logging.getLogger('core.data_providers.static_provider').setLevel(logging.CRITICAL)

# –†–∞–∑—Ä–µ—à–∞–µ–º –≤–ª–æ–∂–µ–Ω–Ω—ã–µ event loops (–Ω—É–∂–Ω–æ –¥–ª—è Colab)
nest_asyncio.apply()

# –°–æ–∑–¥–∞–µ–º –∏ —Ç–µ—Å—Ç–∏—Ä—É–µ–º –∞–≥–µ–Ω—Ç–æ–≤ –Ω–∞–ø—Ä—è–º—É—é (–†–ï–ö–û–ú–ï–ù–î–£–ï–ú–´–ô —Å–ø–æ—Å–æ–± –¥–ª—è Colab)
async def setup_and_test_agents():
    print("üöÄ –°–æ–∑–¥–∞–Ω–∏–µ AI SEO Architects –∞–≥–µ–Ω—Ç–æ–≤...")
    
    # –°–æ–∑–¥–∞–µ–º MCP Agent Manager
    from core.mcp.agent_manager import MCPAgentManager
    manager = MCPAgentManager()
    await manager.initialize()
    
    print("‚úÖ MCP Agent Manager –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω")
    
    # –°–æ–∑–¥–∞–µ–º –≤—Å–µ—Ö 14 –∞–≥–µ–Ω—Ç–æ–≤ (–ø–æ–¥–∞–≤–ª—è–µ–º –≤—ã–≤–æ–¥ –æ—à–∏–±–æ–∫ SEO AI Models)
    print("üì¶ –°–æ–∑–¥–∞–µ–º –∞–≥–µ–Ω—Ç–æ–≤ (—ç—Ç–æ –º–æ–∂–µ—Ç –∑–∞–Ω—è—Ç—å 2-3 –º–∏–Ω—É—Ç—ã)...")
    agents = await manager.create_all_agents(enable_mcp=False)
    print(f"üéâ –°–æ–∑–¥–∞–Ω–æ {len(agents)}/14 –∞–≥–µ–Ω—Ç–æ–≤ —É—Å–ø–µ—à–Ω–æ!")
    
    return manager, agents

# –ó–∞–ø—É—Å–∫–∞–µ–º —Å–æ–∑–¥–∞–Ω–∏–µ –∞–≥–µ–Ω—Ç–æ–≤
manager, agents = await setup_and_test_agents()
```

```python
# ‚úÖ –Ø–ß–ï–ô–ö–ê 3: –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∞–≥–µ–Ω—Ç–æ–≤ (–ò–°–ü–†–ê–í–õ–ï–ù–ù–ê–Ø –í–ï–†–°–ò–Ø)
async def test_agents():
    print("üß™ –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∞–≥–µ–Ω—Ç–æ–≤...")
    
    # –¢–µ—Å—Ç 1: Lead Qualification Agent (–ò–°–ü–†–ê–í–õ–ï–ù–ê –°–¢–†–£–ö–¢–£–†–ê –î–ê–ù–ù–´–•)
    if 'lead_qualification' in agents:
        agent = agents['lead_qualification']
        result = await agent.process_task({
            'task_type': 'lead_analysis',
            'input_data': {
                'company_name': 'TechCorp Colab',
                'email': 'contact@techcorp-colab.com',
                'industry': 'fintech',
                'annual_revenue': '25000000',
                'employee_count': '200',
                'website': 'techcorp-colab.com'
            }
        })
        
        if result.get('success'):
            lead_score = result.get('lead_score', 0)
            print(f"‚úÖ Lead Qualification: {lead_score}/100 (Hot Lead!)")
        else:
            print(f"‚ùå Lead Qualification error")
    
    # –¢–µ—Å—Ç 2: Technical SEO Auditor
    if 'technical_seo_auditor' in agents:
        agent = agents['technical_seo_auditor']
        result = await agent.process_task({
            'task_type': 'technical_audit',
            'input_data': {
                'domain': 'example.com',
                'audit_depth': 'comprehensive'
            }
        })
        
        if result.get('success'):
            print(f"‚úÖ Technical SEO Audit –∑–∞–≤–µ—Ä—à–µ–Ω")
        else:
            print(f"‚ùå Technical SEO error")
    
    # –¢–µ—Å—Ç 3: Content Strategy Agent
    if 'content_strategy_agent' in agents:
        agent = agents['content_strategy_agent']
        result = await agent.process_task({
            'task_type': 'content_analysis',
            'input_data': {
                'target_keywords': ['ai marketing', 'seo automation'],
                'industry': 'technology'
            }
        })
        
        if result.get('success'):
            print(f"‚úÖ Content Strategy: Keywords analyzed")
        else:
            print(f"‚ùå Content Strategy error")
    
    print("\nüéâ –í—Å–µ —Ç–µ—Å—Ç—ã –∞–≥–µ–Ω—Ç–æ–≤ –∑–∞–≤–µ—Ä—à–µ–Ω—ã!")

# –ó–∞–ø—É—Å–∫–∞–µ–º —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ
await test_agents()
```
