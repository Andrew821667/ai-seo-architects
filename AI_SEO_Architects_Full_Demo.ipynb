```python
# üîë –Ø–ß–ï–ô–ö–ê 3: –ù–ê–°–¢–†–û–ô–ö–ê OPENAI API
print("üîë –ù–ê–°–¢–†–û–ô–ö–ê OPENAI API –î–õ–Ø RAG –§–£–ù–ö–¶–ò–ô")
print("=" * 50)

import os
import openai
from openai import OpenAI

# –ü–æ–ª—É—á–∞–µ–º –∫–ª—é—á –∏–∑ —Å–µ–∫—Ä–µ—Ç–æ–≤ Google Colab
api_key = None
api_working = False

print("üîç –ü–æ–∏—Å–∫ OpenAI API –∫–ª—é—á–∞...")

# –°–ø–æ—Å–æ–± 1: –°–µ–∫—Ä–µ—Ç—ã Google Colab (–û–°–ù–û–í–ù–û–ô)
try:
    from google.colab import userdata
    api_key = userdata.get('OPENAI_API_KEY')
    
    if api_key and len(api_key) > 20:
        os.environ["OPENAI_API_KEY"] = api_key
        print(f"‚úÖ –ö–ª—é—á –ø–æ–ª—É—á–µ–Ω –∏–∑ —Å–µ–∫—Ä–µ—Ç–æ–≤ Colab")
        print(f"üî¢ –î–ª–∏–Ω–∞: {len(api_key)} —Å–∏–º–≤–æ–ª–æ–≤")
        print(f"üîê –ü—Ä–µ—Ñ–∏–∫—Å: {api_key[:7]}...")
    else:
        print("‚ùå –ö–ª—é—á –≤ —Å–µ–∫—Ä–µ—Ç–∞—Ö –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –∏–ª–∏ –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç")
        print("\nüìù –ò–ù–°–¢–†–£–ö–¶–ò–Ø –ü–û –ù–ê–°–¢–†–û–ô–ö–ï:")
        print("1. üîë –ù–∞–∂–º–∏—Ç–µ –Ω–∞ –∏–∫–æ–Ω–∫—É –∫–ª—é—á–∞ –≤ –ª–µ–≤–æ–º –º–µ–Ω—é Colab")
        print("2. ‚ûï –î–æ–±–∞–≤—å—Ç–µ –Ω–æ–≤—ã–π —Å–µ–∫—Ä–µ—Ç")
        print("3. üìù Name: OPENAI_API_KEY")
        print("4. üîê Value: –≤–∞—à OpenAI API –∫–ª—é—á (sk-proj...)")
        print("5. üíæ –°–æ—Ö—Ä–∞–Ω–∏—Ç—å –∏ –ø–µ—Ä–µ–∑–∞–ø—É—Å—Ç–∏—Ç—å —ç—Ç—É —è—á–µ–π–∫—É")
        print("\nüåê –ü–æ–ª—É—á–∏—Ç—å –∫–ª—é—á: https://platform.openai.com/api-keys")
        
except ImportError:
    print("‚ö†Ô∏è –ù–µ Google Colab –æ–∫—Ä—É–∂–µ–Ω–∏–µ")
    # –°–ø–æ—Å–æ–± 2: –ü–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –æ–∫—Ä—É–∂–µ–Ω–∏—è
    api_key = os.getenv('OPENAI_API_KEY')
    if api_key:
        print("‚úÖ –ö–ª—é—á –Ω–∞–π–¥–µ–Ω –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –æ–∫—Ä—É–∂–µ–Ω–∏—è")

# –¢–µ—Å—Ç–∏—Ä—É–µ–º API –ø–æ–¥–∫–ª—é—á–µ–Ω–∏–µ
if api_key:
    print("\nüß™ –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï OPENAI API...")
    print("-" * 30)
    
    try:
        # –°–æ–∑–¥–∞–µ–º –∫–ª–∏–µ–Ω—Ç
        client = OpenAI(api_key=api_key)
        
        # –ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π —Ç–µ—Å—Ç
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": "–¢–µ—Å—Ç API. –û—Ç–≤–µ—Ç—å –æ–¥–Ω–∏–º —Å–ª–æ–≤–æ–º: —Ä–∞–±–æ—Ç–∞–µ—Ç?"}],
            max_tokens=10
        )
        
        result = response.choices[0].message.content.strip()
        tokens = response.usage.total_tokens
        
        print(f"‚úÖ OpenAI API —Ä–∞–±–æ—Ç–∞–µ—Ç!")
        print(f"üìù –û—Ç–≤–µ—Ç: '{result}'")
        print(f"üî¢ –¢–æ–∫–µ–Ω—ã: {tokens}")
        print(f"üß† –ú–æ–¥–µ–ª—å: {response.model}")
        
        api_working = True
        
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–∞–±–æ—á–∏–π –∫–ª–∏–µ–Ω—Ç –≥–ª–æ–±–∞–ª—å–Ω–æ
        globals()['OPENAI_CLIENT'] = client
        
    except Exception as e:
        error_msg = str(e)
        print(f"‚ùå –û—à–∏–±–∫–∞ API: {error_msg[:100]}...")
        
        if "insufficient_quota" in error_msg:
            print("üí≥ –ü—Ä–∏—á–∏–Ω–∞: –ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Å—Ä–µ–¥—Å—Ç–≤ –Ω–∞ OpenAI –∞–∫–∫–∞—É–Ω—Ç–µ")
        elif "invalid_api_key" in error_msg:
            print("üîê –ü—Ä–∏—á–∏–Ω–∞: –ù–µ–≤–µ—Ä–Ω—ã–π API –∫–ª—é—á")
        elif "rate_limit" in error_msg:
            print("‚è∞ –ü—Ä–∏—á–∏–Ω–∞: –ü—Ä–µ–≤—ã—à–µ–Ω –ª–∏–º–∏—Ç –∑–∞–ø—Ä–æ—Å–æ–≤ (–∫–ª—é—á –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π)")
            api_working = True  # –ö–ª—é—á —Ä–∞–±–æ—á–∏–π, –ø—Ä–æ—Å—Ç–æ –ª–∏–º–∏—Ç
        else:
            print(f"üîß –ü—Ä–∏—á–∏–Ω–∞: –¢–µ—Ö–Ω–∏—á–µ—Å–∫–∞—è –ø—Ä–æ–±–ª–µ–º–∞")

# –°–æ–∑–¥–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é-–æ–±–µ—Ä—Ç–∫—É –¥–ª—è LLM –∑–∞–ø—Ä–æ—Å–æ–≤
def make_llm_request(prompt, model="gpt-4o-mini", max_tokens=1500):
    """–£–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è LLM –∑–∞–ø—Ä–æ—Å–æ–≤ —Å –æ–±—Ä–∞–±–æ—Ç–∫–æ–π –æ—à–∏–±–æ–∫"""
    if not api_working or 'OPENAI_CLIENT' not in globals():
        return {
            'success': False,
            'content': 'OpenAI API –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω - —Ä–∞–±–æ—Ç–∞–µ–º –≤ –¥–µ–º–æ —Ä–µ–∂–∏–º–µ',
            'demo_mode': True
        }
    
    try:
        client = globals()['OPENAI_CLIENT']
        response = client.chat.completions.create(
            model=model,
            messages=[{"role": "user", "content": prompt}],
            max_tokens=max_tokens,
            temperature=0.3
        )
        
        return {
            'success': True,
            'content': response.choices[0].message.content,
            'tokens': response.usage.total_tokens,
            'model': response.model,
            'demo_mode': False
        }
        
    except Exception as e:
        return {
            'success': False,
            'content': f'–û—à–∏–±–∫–∞ LLM: {str(e)[:100]}',
            'demo_mode': True
        }

# –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –≥–ª–æ–±–∞–ª—å–Ω–æ
globals()['MAKE_LLM_REQUEST'] = make_llm_request

# –§–∏–Ω–∞–ª—å–Ω—ã–π —Å—Ç–∞—Ç—É—Å
print(f"\nüéØ –°–¢–ê–¢–£–° OPENAI:")
print(f"   üîë –ö–ª—é—á –Ω–∞—Å—Ç—Ä–æ–µ–Ω: {'‚úÖ' if api_key else '‚ùå'}")
print(f"   üåê API —Ä–∞–±–æ—Ç–∞–µ—Ç: {'‚úÖ' if api_working else '‚ùå'}")
print(f"   üöÄ –†–µ–∂–∏–º —Ä–∞–±–æ—Ç—ã: {'Production' if api_working else 'Demo'}")

if api_working:
    print("\nüéâ –û–¢–õ–ò–ß–ù–û! OpenAI API –≥–æ—Ç–æ–≤ –¥–ª—è RAG —Ñ—É–Ω–∫—Ü–∏–π")
else:
    print("\n‚ö†Ô∏è –†–∞–±–æ—Ç–∞–µ–º –≤ –¥–µ–º–æ —Ä–µ–∂–∏–º–µ —Å –∫–∞—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–º–∏ –∑–∞–≥–ª—É—à–∫–∞–º–∏")

print("\n‚úÖ OPENAI –ù–ê–°–¢–†–û–ï–ù - –ø–µ—Ä–µ—Ö–æ–¥–∏–º –∫ —Å–æ–∑–¥–∞–Ω–∏—é –∞–≥–µ–Ω—Ç–æ–≤")
```
